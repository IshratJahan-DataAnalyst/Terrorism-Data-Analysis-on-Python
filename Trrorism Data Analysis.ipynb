```python
#Counting the numver of attacks in each country
count_of_attacks=ter['Country'].value_counts()
count_of_attacks
```

```python

attacks = pd.DataFrame({'Country Name':count_of_attacks.index, 'Number Of Attacks':count_of_attacks.values})
# count of years
noy= ter.Year.nunique()
#Here we will calculate the number of attacks for every year in each country
attacks['Attacks Per Year'] = attacks['Number Of Attacks']/noy
attacks.head()
```

```python
pd.crosstab(ter['Year'], ter['Region']).plot(kind='area', figsize=(23,10))
plt.title('Terrorism each year in every Region')
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.barh(ter['Terr_Group'].value_counts()[1:6].index,ter['Terr_Group'].value_counts()[1:6].values)
plt.xlabel('Count of Attacks',fontsize=12)
plt.ylabel('Terrorist Group',fontsize=12)
plt.title('Top5 Active Terrorist Group', fontweight='bold',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
